```python
#importing required libraries
import pandas as pd
import numpy as np
import requests
```

```python
#creating dataframe
df=pd.read_csv('mah_black_spots.csv')
df
```

```python
def calc_dist_from_hosp(origin, destination):
    url = f'https://maps.googleapis.com/maps/api/distancematrix/json?units=metric&origins={origin}&destinations={destination}&key={key}'
    response = requests.get(url)
    data = response.json()
    
    if response.status_code != 200:
        raise ValueError('Failed to fetch data from the Distance Matrix API.')
        distance,duration='<3.10mi','~30min'
    
    if 'rows' not in data or 'elements' not in data['rows'][0]:
        print('No data found!\n')
        distance,duration='<3.10mi','~30min'
#         raise ValueError('Invalid response from the Distance Matrix API.')

    distance = data['rows'][0]['elements'][0].get('distance', {}).get('text')
    duration = data['rows'][0]['elements'][0].get('duration', {}).get('text')
    
    if distance is None or duration is None:
        distance,duration='<3.10mi','~30min'
#         raise ValueError('Invalid distance or duration data in the Distance Matrix API response.')
    
    return distance, duration
```

## Finding all nearby hospitals for Top 10 accident sites

```python
key='AIzaSyCG2EJVnD6OqtU-7qRrHyR-ygXU2CpnEaA'

#search radius
radius=5000

all_lat, all_long, all_names = [], [], []

# Iterate over 'latitude' in the DataFrame
for lat in df['Latitude'][:10]:
    
    # Initialize empty lists for storing data
    hosp_lat, hosp_lng, hosp_sts, hosp_name, hosp_vicinity, distance, time = [], [], [], [], [], [], []
    
    # Get the longitude for the current latitude
    lng = df.loc[df['Latitude'] == lat, 'Longitude'].values[0]
    
    # Create the API request URL
    url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={lat},{lng}&radius={radius}&type=hospital&key={key}"
    
    # Send the request and retrieve the JSON response
    try:
        response = requests.get(url)
        data = response.json()

        # Iterate over each result in the response
        for i in range(len(data['results'])):
            # Check if the hospital is operational or temporarily closed
            if data['results'][i]['business_status'] == 'OPERATIONAL' or data['results'][i]['business_status'] == 'TEMPORARILY_CLOSED':
                # Extract the required data and append to the respective lists
                hosp_lat.append(data['results'][i]['geometry']['location']['lat'])
                hosp_lng.append(data['results'][i]['geometry']['location']['lng'])
                all_lat.append(data['results'][i]['geometry']['location']['lat'])
                all_long.append(data['results'][i]['geometry']['location']['lng'])
                all_names.append(data['results'][i]['name'])
                hosp_sts.append(data['results'][i]['business_status'])
                hosp_name.append(data['results'][i]['name'])
                hosp_vicinity.append(data['results'][i]['vicinity'])
                dist, dur = calc_dist_from_hosp(df.loc[df['Latitude'] == lat, 'Location of accidents\nincluding chainage (km'].values[0], data['results'][i]['vicinity'])
                distance.append(dist)
                time.append(dur)
    except requests.exceptions.RequestException as e:
        print(f'Error occurred while requesting data: {str(e)}')
    
    # Create Series from the collected data
    Latitude = pd.Series(hosp_lat, name='Hospital Latitude')
    Longitude = pd.Series(hosp_lng, name='Hospital Longitude')
    Name = pd.Series(hosp_name, name='Name')
    add = pd.Series(hosp_vicinity, name='Address/Vicinity')
    dist = pd.Series(distance, name='Distance')
    dur = pd.Series(time, name='Duration (from now)')
    
    # Create DataFrame from the Series
    temp = pd.DataFrame({
        'Medical Facility Name': Name,
        'Latitude': Latitude,
        'Longitude': Longitude,
        'Address/Vicinity': add,
        'Distance': dist,
        'Duration (from now)': dur
    })
    
    # Print the accident location and display the DataFrame
    print('For Accident Location: ' + df.loc[df['Latitude'] == lat, 'Location of accidents\nincluding chainage (km'].values[0])
    print()
    display(temp)
    print()
```

```python
df1=pd.DataFrame({
    'Name':all_names,
    'Latitude':all_lat,
    'Longitudes':all_long
})
df1
```
