```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('only_road_accidents_data3.csv')
data.head(20)
```

```python
data['STATE/UT'].value_counts()
```

```python
df=data[data['STATE/UT']=='Maharashtra']
df
```

```python
plt.plot('YEAR','Total',marker='o',data=df)
plt.xlabel('YEAR OF ACCIDENT')
plt.ylabel('NUMBER OF CASUALTIES')
plt.title('Yearwise trend of accidents')
plt.grid()
```

```python
main=df.columns[2:10]
```

```python
import matplotlib.pyplot as plt

# Data for Maharashtra for each year
years = range(2001, 2015)
main = df.columns[2:10]

# Create subplots
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))

# Iterate over each year
for i, year in enumerate(years):
    # Get the values for the current year
    values = []
    for column in main:
        values.append(df[df['YEAR'] == year][column].values[0])
    
    # Determine the subplot position
    row = i // 4
    col = i % 4
    
    # Create a horizontal bar plot for the current year
    axs[row, col].barh(main, values)
    axs[row, col].set_title(str(year))
    
    # Invert the y-axis
    axs[row, col].invert_yaxis()

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Maharashtra for each year (replace with your actual data)
years = range(2001, 2015)

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Define the time segments for x-axis grouping
time_segments = main

# Set the width of each bar
bar_width = 1.5 / len(main)

# Iterate over each year
for year in years:
    # Initialize the bar positions
    bar_positions = np.arange(len(time_segments))

    # Get the values for the current year
    values = [df[df['YEAR'] == year][column].values[0] for column in main]

    # Plot the bars for the current year
    ax.barh(bar_positions, values, height=bar_width, label=str(year))

# Set the y-axis tick positions and labels
ax.set_yticks(np.arange(len(time_segments)))
ax.set_yticklabels(time_segments)

# Set the x-axis and y-axis labels
ax.set_xlabel('Value')
ax.set_ylabel('Time Segments')

# Set the title
ax.set_title('Trends for Maharashtra (2001-2014)')

# Add a legend
ax.legend(title='Year')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for Maharashtra for each year
years = range(2001, 2015)
main = df.columns[2:10]

# Create a line plot
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate over each category
for column in main:
    # Get the values for each year
    values = [df[df['YEAR'] == year][column].values[0] for year in years]
    
    # Plot the line for the current category
    ax.plot(years, values, marker='o', label=column)

# Set the x-axis and y-axis labels
ax.set_xlabel('Year')
ax.set_ylabel('Value')

# Set the title
ax.set_title('Trends for Maharashtra (2001-2014)')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
